## To Create the Decision Tree classifier and visualize it graphically

Dataset https://bit.ly/3kXTdox

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("Iris.csv")
```

```python
#Vissualizing Data
data.head()
```

```python
y=np.array(data.Species).reshape(-1,1)
```

```python
data['Target']=list((LabelEncoder().fit_transform(y)).ravel())
```

```python
data.head()
```

```python
X=data[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 
y=data.Target
```

```python
#Initializing the Classifier Instance
clf=tree.DecisionTreeClassifier(random_state=0)

# Fitting the tree with the data
clf.fit(X,y)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True,class_names=data.Species.value_counts().index.tolist(),feature_names=X.columns,)
```
